```python
import pandas as pd
import numpy as np
import os
import sys
```

```python
# Get list of files to process
path = 'C:/Master_rosters'

files = os.listdir(path)
```

```python
# Create empty dataframe to hold all the data
master_df = pd.DataFrame(columns=['Last Name', 'First Name', 'Class Working As', 'Hospital Assignment', 'Date'])
```

```python
for f in files:
    # read the file to a dataframe, select the first sheet automatically
    print(f)
    try:
        file_df = pd.read_excel(f, sheet_name='Master Roster')
    except: 
        try:
            file_df = pd.read_excel(f, sheet_name='Sheet1')
        except:
            try:
                file_df = pd.read_excel(f, sheet_name='Live Roster')
            except:
                print("Last File Failed to find proper Sheet: ")
                file_df = pd.read_excel(f)
    
    # get the date
    date = f[:10]
    year = date[6:10]
    month = date[:2]
    day = date[3:5]
    
    final_date = year + "/" + month + "/" + day
    
    # rename the hospital and class columns if not standard
    try:
        file_df.rename(columns= {'Hospital':'Hospital Assignment'}, inplace=True)
    except:
        continue
    
    try:
        file_df.rename(columns= {'AS':'Hospital Assignment'}, inplace=True)
    except:
        continue
    
    try:
        file_df.rename(columns= {'Facility ':'Hospital Assignment'}, inplace=True)
    except:
        continue
    
    try:
        file_df.rename(columns= {'Facility':'Hospital Assignment'}, inplace=True)
    except:
        continue
    
    try:
        file_df.rename(columns={'Class':'Class Working As'}, inplace=True)
    except:
        try:
            file_df.rename(columns={'Class':'Class Working As'}, inplace=True)
        continue
        
    # create temp dataframe with all necessary columns
    holding_df = file_df[['Last Name', 'First Name', 'Class Working As', 'Hospital Assignment']].copy()
    holding_df['Date'] = final_date
    
    # append temp dataframe to master
    master_df.append(holding_df, ignore_index=True)
    
```
