Compile Payroll and Alert Media files to determine who to delete

```python
import pandas as pd
import numpy as np
import os
import sys
```

```python
# Get list of files to process - Master rosters
path = 'C:/DataCleanup/Data'
```

```python
# Compare Alert Media to already worked Payroll report.
try:
    payroll2020_df = pd.read_excel(path + '/Payroll report.xlsx')
except: 
    print('Cannot open Payroll file')
    
payroll2020_df
```

```python
# Get rid of rows where 'PD 2020' = N
payroll_yes_df = payroll2020_df[payroll2020_df['PD 2020'] == 'Y']
payroll_yes_df
```

```python
try:
    payroll20_21_df = pd.read_excel(path + '/Payroll report.xlsx', sheet_name='Paid 2020 & 2021')
except: 
    print('Cannot open Payroll file')
    
payroll20_21_df
```

```python
try:
    payroll20_df = pd.read_excel(path + '/Payroll report.xlsx', sheet_name='Paid 2020')
except: 
    print('Cannot open Payroll file')
    
payroll20_df
```

```python
try:
    payroll21_df = pd.read_excel(path + '/Payroll report.xlsx', sheet_name='Paid 2021')
except: 
    print('Cannot open Payroll file')
    
payroll21_df
```

```python
# add final payroll file
try:
    payroll_workday_df = pd.read_excel(path + '/workdayData.xlsx')
except: 
    print(path + 'Cannot open Payroll file')
    
payroll_workday_df
```

```python
# Combine 2020 and 2021 to get full list of those who have been paid.  
payroll_df1 = payroll_yes_df.append(payroll20_21_df, ignore_index=True)
payroll_df2 = payroll_df1.append(payroll20_df, ignore_index=True)
payroll_df3 = payroll_df2.append(payroll21_df, ignore_index=True)
```

```python
sorted_payroll_df = payroll_df3.sort_values(by=['NTNL'])
sorted_payroll_df
```

```python
unique_payroll_df = sorted_payroll_df.drop_duplicates(subset=['NTNL'], ignore_index=True)
```

```python
# set the index to Email after renaming the column
unique_payroll_df.rename(columns={'EMAIL':'Email'}, inplace=True)
payroll_email_df = unique_payroll_df[['Email', 'FULL NAME']]
payroll_email_df
```

```python
# Add the Workday records to get to the final list
final_payroll_df = payroll_email_df.append(payroll_workday_df, ignore_index=True)

final_payroll_df

# save to file
final_payroll_df.to_csv("full_payroll.csv")
```
