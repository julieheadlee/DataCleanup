```python
import pandas as pd
import numpy as np
import os
import sys
from datetime import date
```

<h1>Open Current Cleanup Master

```python
master_path = 'C:/DataCleanup/Data/'
master_file = 'Master Tracking Sheet.xlsx'
file_df = pd.read_excel(master_path + master_file, dtype={'ProfileID_M': str})
file_df
```

<h1>Open Alert Media download

```python
am_path = 'C:/DataCleanup/Data/AM Files/'
am_file = 'AM download.csv'
am_df = pd.read_csv(am_path + am_file, low_memory=False)
am_df
```

```python
#What types are the IDs
dataTypeMobilize = dict(file_df.dtypes)
print(dataTypeMobilize)

dataTypeAM = dict(am_df.dtypes)
print(dataTypeAM)
```

```python
# Do an inner join to find all records in both AM and Mobilize by Profile ID

am_master_merge_df = pd.merge(file_df, am_df, left_on='ProfileID_M', right_on='employee_id', 
                              how='inner', suffixes=('_master', '_am'))
```

```python
# Delete all rows with no email 

file_df.dropna(subset=['email_M'], inplace=True)
file_df
```

```python
am_master_merge_email_df = pd.merge(file_df, am_df, left_on='email_M', right_on='email', how='inner', suffixes=('_master', '_am'))
```

```python
am_master_merge_df
```

```python
am_master_merge_email_df
```

```python
am_master_merge_df.to_excel("Mobilize_AM_byID.xlsx" )
```

```python
am_master_merge_email_df.to_csv("Mobilize_AM_byEmail.csv")
```

```python
# Create separate dataframe for the duplicate emails
am_master_merge_email_df['Duplicate?'] = am_master_merge_email_df.duplicated(subset=['email_M'], keep=False)
```

```python
am_master_merge_email_df
```

```python
email_dups_df = am_master_merge_email_df.loc[am_master_merge_email_df['Duplicate?'] == True]
```

```python
email_dups_df.to_csv("duplicate_emails.csv")
```

```python
email_non_dups_df = am_master_merge_email_df.loc[am_master_merge_email_df['Duplicate?'] == False]
email_non_dups_df.to_csv('unique email records.csv')
```

```python
# Need three tracking sheets - those whose emails and Profile ID's match, those whose emails match, and those 
# whose Profile ID's match.
am_master_merge_email_df['Full Match?'] = am_master_merge_email_df.ProfileID_M.isin(am_master_merge_df.ProfileID_M)
merge_email_no_match_df = am_master_merge_email_df[am_master_merge_email_df['Full Match?'] == False]
```

```python
am_master_merge_df['Full Match?'] = am_master_merge_df.ProfileID_M.isin(am_master_merge_email_df.ProfileID_M)
merge_id_no_match_df = am_master_merge_df[am_master_merge_df['Full Match?'] == False]
```

```python
both_df = am_master_merge_df.merge(am_master_merge_email_df, on='ProfileID_M', how='inner')
```

```python
merge_email_no_match_df.to_csv("Emails no match in ID.csv")
merge_id_no_match_df.to_csv("IDs no match in emails.csv")
both_df.to_csv("Full ID and Email match.csv")
```
